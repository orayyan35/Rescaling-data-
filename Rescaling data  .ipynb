```python
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1000, 1500, 2000, 2500, 3000]  # House areas
y = [300000, 450000, 600000, 750000, 900000]  # House prices
```

```python
x_min = min(x)
x_max = max(x)
y_min = min(y)
y_max = max(y)

rescaled_x = [(x_i - x_min) / (x_max - x_min) for x_i in x]
rescaled_y = [(y_i - y_min) / (y_max - y_min) for y_i in y]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x, y, c='blue', label='Original Data')
plt.xlabel('House Area')
plt.ylabel('House Price')
plt.title('Original Data')
plt.legend()
plt.subplot(1, 2, 2)
plt.scatter(rescaled_x, rescaled_y, c='red', label='Rescaled Data')
plt.xlabel('Rescaled House Area')
plt.ylabel('Rescaled House Price')
plt.title('Rescaled Data')
plt.legend()
plt.tight_layout()
plt.show()
```
